```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, "r--")
```

```python
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("Title")
```

```python
#OO

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.4,0.5,0.3 ])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots(nrows = 2, ncols = 1)

axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(y,x)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

```python
fig = plt.figure(figsize = (8,2))

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize = (8,5))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig("my_pic.png",dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label="X Square")
ax.plot(x,x**3, label = "X Cubed")


ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color= 'purple', lw = 0.5, ls= '-', marker = 'o', markersize = 10,
       markerfacecolor = 'yellow', markeredgewidth=3,markeredgecolor = "green")
```

```python

import numpy as np 
x = np.arange(0,100)

y = x *2

z = x** 2
```

```python
x
y
z
```

```python
y
```

```python
z
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)

ax.set_xlabel("X")

ax.set_ylabel("Y")

ax.set_title("Title")
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])

ax1.plot(x,y,color="red")
ax2.plot(x,y,color="blue")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax.plot(x,z)
ax.set_xlabel("X")
ax.set_ylabel("Z")

ax2.plot(x,y)
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_title("Zoom")
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig,axes = plt.subplots(1,2, figsize=(12,2))

axes[0].plot(x,y,color = 'red',ls = '--',lw=3)
axes[1].plot(x,z,color = 'blue', lw = 3)
```
